```python
f = open("in.txt").read().strip().split("\n")
lines = [line for line in f]

ans = 0
cur = 50
for ins in lines:
    d = ins[0]
    num = int(ins[1:])
    if num > 100:
        ans += num // 100
        num %= 100
    count = False if not cur else True
    if d == 'R':
        cur += num
    else:
        cur -= num
    
    if cur >= 100:
        ans += 1 * count
        cur -= 100
    elif cur <= 0:
        ans += 1 * count
        cur = (100 + cur) % 100
ans
# 2858
```
