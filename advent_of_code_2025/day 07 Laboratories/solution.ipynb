```python
from aocutils.grid import gridneigh, dimensions
from collections import deque
from functools import cache
# grid
grid, neigh = gridneigh("in.txt", to_dict=True, parser=lambda x: [(ch) for ch in x])
(minr, minc), (maxr, maxc), (rows, cols) = dimensions(grid)
```

```python
splitters = {k for k,v in grid.items() if v == '^'}
start = [k for k,v in grid.items() if v == 'S'].pop()

tocheck = [start]
seen = set()
ans = 0
while tocheck:
    cur = tocheck.pop()
    new = cur[0]+1, cur[1]
    if new in grid and new not in splitters:
        tocheck.append(new)
    if new in splitters and new not in seen:
        seen.add(new)
        ans += 1
        new1 = (new[0], new[1]-1)
        new2 = (new[0], new[1]+1)
        for new in [new1, new2]:
            if new in grid and new not in seen:
                tocheck.append(new)
                seen.add(new)
ans
```

```python
@cache
def dp(r,c):
    if r == 0:
        if c == start[1]:
            return 1    
        else:
            return 0
    nr = r - 1
    ways = 0
    if (nr, c + 1) in splitters:
        ways += dp(nr, c + 1)
    if (nr, c - 1) in splitters:
        ways += dp(nr, c - 1)
    if (nr, c) in grid and (nr, c) not in splitters:
        ways += dp(nr, c) 
    return ways
ans = 0
for c in range(minc, maxc+1):
    ans += dp(maxr, c)
ans
```
