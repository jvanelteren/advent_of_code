```python
from aocutils.common import ints
from aocutils.grid import dimensions
from collections import defaultdict

f = open("in.txt").read().strip().split("\n")
lines = [line for line in f]

reds = []
greens = set()

rows = defaultdict(set)
cols = defaultdict(set)
cur = ints(lines[-1])

# get all the reds and the greens in between on the lines
# also build row and column maps, but exclude the last red
for line in lines:
    red = ints(line)
    reds.append(ints(line))
    if red[0] == cur[0]:
        for y in range(min(red[1], cur[1]), max(red[1], cur[1]) + 1):
            greens.add((red[0], y))
            cols[y].add((red[0]))
        cols[y].remove(red[0])
    if red[1] == cur[1]:
        for x in range(min(red[0], cur[0]), max(red[0], cur[0]) + 1):
            greens.add((x, red[1]))
            rows[x].add(red[1])
        rows[x].remove(red[1])
    cur = red
    
for k in sorted(rows):
    rows[k] = sorted(rows[k])
for k in sorted(cols):
    cols[k] = sorted(cols[k])

d = dimensions(reds)
d
```

```python
import bisect

# caching doesn't help here, even makes it much slower
g, b = 0,0
def valid(x,y):
    if (x,y) in greens:
        global g
        g += 1
        return True
    global b
    b += 1
    return bisect.bisect_right(cols[y], x) % 2 == 1 # you could also do it with the cols

def all_points_on_square_edge(l1,l2):
    x1, y1 = l1
    x2, y2 = l2
    for x in range(min(x1, x2), max(x1, x2) + 1):
        yield(x,y1)
        yield(x,y2)
    for y in range(min(y1, y2), max(y1, y2) + 1):
        yield(x1,y)
        yield(x2,y)
    
def opposite_corners(l1,l2):
    x1, y1 = l1
    x2, y2 = l2
    yield(x1,y2)
    yield(x2,y1)
    
ans = 0
for idx, l1 in enumerate(reds):
    for l2 in reds[idx+1:]:
        size = (abs(l1[0] - l2[0]) + 1) * (abs(l1[1] - l2[1]) + 1)
        # first do the easy checks
        if size > ans:
            if all(valid(*cell) for cell in opposite_corners(l1, l2)):
                # the slow check
                if all(valid(*cell) for cell in all_points_on_square_edge(l1, l2)):
                    ans = size
                    print('new',l1,l2)
ans
```
