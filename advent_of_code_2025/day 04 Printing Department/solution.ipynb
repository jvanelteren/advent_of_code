```python
from aocutils.grid import gridneigh, neighbors

# grid
grid, neigh = gridneigh("in.txt", to_dict=True, parser=lambda x: [(ch) for ch in x])

ans = 0
removed = True
while removed:
    removed = False    
    for cell in list(grid.keys()): # list to avoid runtime error due to changing dict size during iteration
        paper = 0
        for n in neighbors(cell, diag=True):
            if n in grid and grid[n] == '@':
                paper +=1
        if paper < 4 and grid[cell] == '@':
            ans +=1
            del grid[cell] # could also set to '.' but this is faster in runtime. Less elegant though
            removed = True
ans
    
        
```
