```python
@cache
def rotate(locs, flip=True):
    
    """
        Rotates a given pattern by 90 degrees (like a puzzle piece)
        when flip=True, also piece around leading to maximum of 8 options
    """
    def normalize(locs):
        minr = min(r for r,c in locs)
        minc = min(c for r,c in locs)
        return {(r-minr, c-minc) for r,c in locs}

    options = [{(r,c) for r,c in locs},
            {(c,-r) for r,c in locs},
            {(-r,-c) for r,c in locs},
            {(-c,r) for r,c in locs},
    ]
    assert len(options[0]) == len(locs)
    if flip:
        options += [
            {(-r,c) for r,c in locs},
            {(c,r) for r,c in locs},
            {(r,-c) for r,c in locs},
            {(-c,-r) for r,c in locs}]
    return [normalize(o) for o in options]
    
    
class Piece():
    def __init__(self, locs:set, color='red'):
        self.options = rotate(tuple(locs), flip=True)
        self.color = color
    
    def size(self):
        return len(self.options[0])
    
    def alloptions(self):
        for o in self.options:
            yield o
    
    
    
l = set([(0,0),(1,0),(1,1)])
from heapq import heapify, heappop, heappush
class Board():
    def __init__(self, rows,cols, spots=None) -> None:
        self.rows, self.cols = rows, cols
        self.spots = {(r,c) for r in range(rows) for c in range(cols)} if not spots else spots
        self.played = set()
    def isfull(self):
        return False if self.spots else True
    
    def spotsfree(self):
        return len(self.spots)
    
    def nextfree(self):
        return min(self.spots)
            
    def place(self, piece, loc, color):
        # return True if placed, false if not able to place
        dr,dc = loc
        take = {(r+dr, c+dc) for r,c in piece}
        if take - self.spots:
            return False
        else:
            self.spots -= take
            self.played.add((color, frozenset(take)))
            return True
        
    def remove(self, piece, loc, color):
        # return True if placed, false if not able to place
        dr,dc = loc
        free = {(r+dr, c+dc) for r,c in piece}
        if free & self.spots:
            return False
        else:
            self.spots |= free
            self.played.remove((color, frozenset(free)))
            return True
        
p = Piece(l)

class Pieces():
    def __init__(self, piecelist, types) -> None:
        
        self.pieces = piecelist # but not a list of piece objects, but a list op number of pieces of each type
        self.placed = [0] * len(piecelist)
        self.types = types
    
    def allplayable(self):
        for i in range(len(self.pieces)):
            if self.placed[i] < self.pieces[i]:
                self.placed[i] += 1
                yield self.types[i]
                self.placed[i] -= 1
                
                
    
```
