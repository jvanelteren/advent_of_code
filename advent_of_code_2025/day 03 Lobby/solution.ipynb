```python
# p1
lines = open("in.txt").read().strip().split("\n")
ans = 0
for line in lines:
    for num in range(99,0,-1):
        ileft = line.find(str(num)[0])
        iright = line.rfind(str(num)[1])
        if ileft != -1 and iright != -1 and ileft < iright:
            ans += num
            break
    else:
        print('not found')
ans
    
```

```python
from functools import cache
@cache
def dp(num, togo, idx):
    if idx == len(num) or togo == 0:
        return '0'
    else:
        take = num[idx] + dp(num, togo - 1, idx + 1)
        skip = dp(num, togo, idx + 1)
        return take if int(take) > int(skip) else skip
        # could also have avoid the int conversion by comparing lengths first
        return max((take, skip), key=lambda x: (len(x), x))
sum(int(dp(line, 12,0)[:-1]) for line in lines)
```

```python
from functools import cache
@cache
def dp(num, togo, idx):
    if togo == 0:
        return ""
    if idx == len(num):
        return ""  # impossible path: too short
    else:
        take = num[idx] + dp(num, togo - 1, idx + 1)
        skip = dp(num, togo, idx + 1)
        # could also have avoid the int conversion by comparing lengths first
        return max((take, skip), key=lambda x: (len(x), x))
sum(int(dp(line, 12,0)) for line in lines)
```
