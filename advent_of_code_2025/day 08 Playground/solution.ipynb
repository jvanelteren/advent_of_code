```python
from math import sqrt
from aocutils.common import ints
from aocutils.special import UnionFind
from itertools import product

f = open("in.txt").read().strip().split("\n")
boxes = [ints(line) for line in f]

eucl = lambda a,b,c,x,y,z: sqrt((a - x)**2 + (b - y)**2 + (c - z)**2)
distances = sorted([(eucl(*a, *b),a,b) for a,b in product(boxes, boxes) if a < b])

uf = UnionFind(boxes)
seen = set()
added = 0

for dist, a, b in distances:
    if (a,b) not in seen:
        uf.union(a,b)
        seen.add((a,b))
        added += 1
        if len(uf.groups()) == 1:
            print(f'all connected p1 after {added} iterations, answer: {a[0] * b[0]}')
            break
        if added == 1000:
            l = sorted([len(g) for g in uf.groups()], reverse=True)
            print('p1 answer after 1000 iterations ', l[0] * l[1] * l[2])
```
