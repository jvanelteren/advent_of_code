```python
def get_numbers(binary):
    num = ''
    for i in range(6, len(binary), 5):
        first = binary[i]
        num += binary[i+1:i+5]
        if first != '1':
            # this was the last number
            break
    return (int(num,2), i+5)


def parse_until_len(binary):
    start = 7 + 15
    stop = int(binary[7:start],2)
    print('total len bits',stop)
    pnt = 0
    answers = []
    while pnt < stop:
        res, l = parse_bin(binary[start + pnt : start + stop])
        print('res received', res, l)
        answers.append(res)
        pnt += l
    pnt = start + stop
    return answers, pnt


def parse_until_npackets(binary):
    npackets = int(binary[7:7+11],2)
    packets = 0
    pnt = 7+11
    answers = []
    while packets < npackets:
        res, subpnt = parse_bin(binary[pnt:])
        pnt += subpnt
        packets += 1
        answers.append(res)
    return answers, pnt


def parse_bin(binary):
    version = int(binary[:3], 2)
    typeid = int(binary[3:6], 2)
    
    if typeid == 4:
        return get_numbers(binary)

    else:
        length_typeid = binary[6]
        if length_typeid == '0':
            print('operator packet with length')
            answers, pnt = parse_until_len(binary)
        elif length_typeid == '1':
            print('operator packet with packets expected')
            answers, pnt = parse_until_npackets(binary)
        else:
            print('other length_typeid not implemented')
            sys.exit()
                
        match typeid:
            case 0:
                res = sum(answers)
            case 1:
                res = math.prod(answers)
            case 2:
                res = min(answers)
            case 3:
                res = max(answers)
            case 5:
                res = 1 if answers[0] > answers[1] else 0
            case 6:
                res = 1 if answers[0] < answers[1] else 0
            case 7:
                res = 1 if answers[0] == answers[1] else 0
        return res, pnt
```

```python
def is_palindrome(num):
    # Skip single-digit inputs
    if num // 10 == 0:
        return False
    temp = num
    reversed_num = 0

    while temp != 0:
        reversed_num = (reversed_num * 10) + (temp % 10)
        temp = temp // 10

    if num == reversed_num:
        return True
    else:
        return False
```

```python
def infinite_palindromes():
    num = 0
    while True:
        if is_palindrome(num):
            i = (yield num)
            if i is not None:
                num = i
        num += 1
```
