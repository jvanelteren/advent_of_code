```python
# spend too much time on this and the next one, works with the test cases but not with a trillion options...
f=open('test0.txt')
lines = [int(line.rstrip('\n')) for line in f]
lines.sort()
lines.insert(0,0)
lines.append(lines[-1]+3)

n_items=len(lines)

import aoc
def find_sol(p):
    val = lines[p]
    sol = []
    for i in range(1,4):
        if p+i < n_items:
            if p+i == n_items -1:
                if lines[p+i]-val <4:
                    return 'found'
            elif lines[p+i]-val < 4:
                sol.append(find_sol(p+i))
    return sol
len(aoc.flatten(find_sol(0)))
```

```python
from networkx.algorithms.simple_paths import all_simple_paths
import networkx as nx

f=open('test1.txt')
lines = [int(line.rstrip('\n')) for line in f]
lines.sort()
lines.insert(0,0)
lines.append(lines[-1]+3)

G = nx.DiGraph()
for i in lines:
    for j in range(1,4):
        if i+j in lines: G.add_edge(i,i+j)

sum(1 for path in all_simple_paths(G, lines[0], lines[-1]))
    
```
