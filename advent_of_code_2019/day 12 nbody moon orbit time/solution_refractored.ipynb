```
def getdelta(x,y):
  res = y['pos']-x['pos']
  res = (res>0)*1+(res<0)*-1
  return res

def compare(x,y):
  for i in range(len(x)):
    for key in ['pos','vel']:
      if any(x[i][key]!=y[i][key]):
        return False
  return True

bodies = [[14,4,5],[12,10,8],[1,7,-10],[16,-5,3]] #input

g = []
for b in bodies:
  planet = {
      'pos':np.array(b),
      'vel':np.array([0,0,0]),
      'pot':sum(abs(np.array(b))),
      'kin':0,
      'tot':0,
  }
  g.append(planet)
start = deepcopy(g)

for i in range(0,1000):
  # apply gravity
  for x in g:
    for y in g:
      x['vel'] += getdelta(x,y)
  # apply velocity and calculate energies
  for x in g:
    x['pos']+=x['vel']
    x['pot']= sum(abs(x['pos']))
    x['kin']= sum(abs(x['vel']))
    x['tot']= x['pot']*x['kin']

  if sum([x['tot'] for x in g])==0:
    if compare(start,g): print ('found',i)

print(sum([x['tot'] for x in g])) # for puzzle 1 after 1000 runs
```

```
# part 2
def delta(x,positions):
  res = positions-x
  res = (res>0)*1+(res<0)*-1
  # print (res)
  return sum(res)

def step(p):
  v = np.array([0,0,0,0])
  start = deepcopy(p)
  #update gravity
  counter = 0
  while True:
    counter += 1
    for i in range(4):
      v[i] = v[i] + delta(p[i],p)
    for i in range(4):
      p[i]+= v[i]
    # update position
    if all(p==start) and all(v==np.array([0,0,0,0])): return counter

bodies = [[14,4,5],[12,10,8],[1,7,-10],[16,-5,3]] #input
#bodies = np.array([[-1.,0,2],[2,-10,-7],[4,-8,8],[3,5,-1]]) #ex1
#bodies = np.array([[-1.,0,2],[2,-10,-7],[4,-8,8],[3,5,-1]]) #ex2
res = np.array([step(np.array([b[i] for b in bodies]))for i in range(3)])
res 
```

```
res = res.astype(np.int64)

def lcm(a, b):
    """Compute the lowest common multiple of a and b"""
    return a * b / gcd(a, b)
int(lcm(int(lcm(res[0],res[1])),res[2]))
```
