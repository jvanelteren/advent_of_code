```python
from IPython.core.debugger import set_trace
import numpy as np
import sys
import collections
import re
f=open('day17input.txt') #not with read because thats probably the whole file
lines = [line.rstrip('\n') for line in f]
```

```python
trog =[]
x1=[]
y1=[]
for i,line in enumerate(lines):
    #print (i,line)
    print (line.split(','))
    if line.split(',')[0][0]=='x':
        resultx = re.search('\=(.*)',line.split(',')[0])[1]
        resulty = re.search('\=(.*)',line.split(',')[1])[1]
    else:
        resulty = re.search('\=(.*)',line.split(',')[0])[1]
        resultx = re.search('\=(.*)',line.split(',')[1])[1]
    
    if resultx.find("..")>-1:
        slice0 = int(resultx.split('..')[0])
        slice1 = int(resultx.split('..')[1])+1
        resultx = slice(slice0,slice1)
        x1.append(slice0)
        x1.append(slice1)
    else: 
        resultx = int(resultx)
        x1.append(resultx)

    if resulty.find("..")>-1:
        slice0 = int(resulty.split('..')[0])
        slice1 = int(resulty.split('..')[1])+1
        resulty = slice(slice0,slice1)
        y1.append(slice0)
        y1.append(slice1)
    else: 
        resulty = int(resulty)
        y1.append(resulty)
        


    trog.append((resulty,resultx))

    
print (trog)    
    
    #trog.append([int(item) for item in result.split(',')])


#trog = np.array(trog)
```

```python
print (min(x1))
print (max(x1))
print (min(y1))
print (max(y1))

```

```python
a = np.chararray(unicode=True,shape=(max(y1),max(x1)+1))
a.fill('.')
print (a.shape)
for line in trog:
    #print (line)
    a[line[0],line[1]]='#'
b= (a[min(y1)-1:max(y1),min(x1):max(x1)])

a[1,500]='|'
solids = ['#','~']
todo =[]
def inc_vertical(x,y):
    global a,todo

    a[x,y]='|'
    if x < a.shape[0]-1:
        if a[x+1,y] in solids: #if bottom is solid then check if horizontal expand is possible
            #print ('call hor expand',x+1,y)
            inc_horizontal(x,y)
        else: #bottom is nonsolid, so fall one deeper and call self
            a[x+1,y]='|'
            if x+1<a.shape[0]:
                
                inc_vertical(x+1,y)

def inc_horizontal(x,y):
    global a,todo
    #print ('horizontal',x,y)
    if a[x,y+1] not in solids:
        rightsolid = inc_horizontal_right(x,y+1)
    else:
        rightsolid = True
    if a[x,y-1] not in solids:
        #print ('call hor left expand',x,y-1)
        leftsolid = inc_horizontal_left(x,y-1)
    else:
        leftsolid = True

        
    if rightsolid and leftsolid:
        a[x,y]='~'
        #print (x-1,y)

  
        if a[x-1,y]=='|':todo.append((x-1,y))
        changesolidwaterleft(x,y-1)
        changesolidwaterright(x,y+1)

def changesolidwaterleft(x,y):
    global a,todo

    if a[x,y] not in solids:
        a[x,y]='~'
        if a[x-1,y]=='|':todo.append((x-1,y))
        changesolidwaterleft(x,y-1)
        
def changesolidwaterright(x,y):
    global a,todo
    if a[x,y] not in solids:
        a[x,y]='~'
        if a[x-1,y]=='|':todo.append((x-1,y))
        changesolidwaterright(x,y+1)
    
    
        
        
def inc_horizontal_right(x,y):
    global a,todo
    a[x,y]='|'
    if a[x+1,y] not in solids:#chck if bottom is not open
        ###inc_vertical(x+1,y)
        todo.append((x+1,y))
        return False
    elif a[x,y+1] in solids:
        return True
    
    else: return inc_horizontal_right(x,y+1)

def inc_horizontal_left(x,y):
    global a, todo
    #print ('catch hor left expand',x,y)
    a[x,y]='|'
    if a[x+1,y] not in solids:#chck if bottom is not open
        #print ('bottomopen')
        ###inc_vertical(x+1,y)
        todo.append((x+1,y))
        return False
    elif a[x,y-1] in solids:#stop is left is closed
        return True
    
    else: 
        #print ('recall hor left expand',x,y-1)
        return inc_horizontal_left(x,y-1)    

def all_falling_water():
    global a, todo
    todocopy = todo.copy()
    todo=[]
    print (todocopy)
    for i in todocopy:
        #print ('i',i)
        if i[0]<a.shape[0]:
            inc_vertical(i[0],i[1])
            #inc_horizontal(xs[i],ys[i])
    
    return 

xs, ys = np.nonzero(a =='|')
#print (a.shape[0])
for i in range(len(xs)):
    if xs[i]<a.shape[0]:
        inc_vertical(xs[i],ys[i])


total = 0

def helpme():
    global total
    global a
    for x in range(100000):
        all_falling_water()
        if x %100==0:
            flowing = np.sum(a.count('|'))
            still = np.sum(a.count('~'))
            print (flowing+still,flowing,still)
            new = flowing+still
            if total == new:
                sys.exit()
                #print (total)
            else: total = new
            
%load_ext line_profiler           
%lprun -f all_falling_water helpme()
```
