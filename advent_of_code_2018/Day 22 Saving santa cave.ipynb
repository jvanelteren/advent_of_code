```python
import numpy as np
import collections
import networkx as nx
```

```python
#target = (10,10)
#depth = 510

target = (721,7)
depth = 9171
caveshape = (1500,70)

xmultiply= 48271
ymultiply = 16807
m = 20183
endm = 3

cave = np.zeros(shape=caveshape)
erosion = np.zeros(shape=cave.shape)
geology = np.zeros(shape=cave.shape)

cave.astype(np.int64)
for y in range(cave.shape[1]): #geology index colums
    cave[0,y]=(ymultiply*y)%m
    erosion[0,y]=((cave[0,y]+depth)%m)
for x in range(cave.shape[0]):#geology index rows
    cave[x,0]=(xmultiply*x)%m
    erosion[x,0]=((cave[x,0]+depth)%m)

for x in range(1,cave.shape[0]): #geology index rest
    for y in range(1,cave.shape[1]):
        if x==target[0] and y==target[1]:
            cave[target]=0
            erosion[target]=depth%m
        else:
            cave[x,y]=(erosion[x-1,y]*erosion[x,y-1])%m
            erosion[x,y]=((cave[x,y]+depth)%m)
        
for i, cell in np.ndenumerate(geology):
    geology[i]=erosion[i]%endm
    
```

```python
neighborarray = np.array([(0,1),(1,0)])

def getneighbors(cell):
    neighbors = cell+neighborarray
    return [tuple(x) for x in neighbors if (x[0]>=0 and x[1]>=0 and x[0]<geology.shape[0] and x[1]<geology.shape[1])]


gear={0:['t','c'],
     1:['c','n'],
     2:['t','n']}
edges = []
for i, cell in np.ndenumerate(geology):
    node1 = (i,gear[cell][0])
    node2 = (i,gear[cell][1])
    neighbors=getneighbors(i)
    
    for owngear in gear[geology[i]]:
        for n in neighbors:
            for neighborgear in gear[geology[n]]:
                if owngear == neighborgear:
                    weight = 1
                #else:
                #    weight = 8
                    edges.append(((i,owngear),(n,neighborgear),weight))
                #print (edges[-1])
    edges.append((node1,node2,7))
    #print (edges[-1])

#puzzle2

G=nx.Graph()
#G.add_nodes_from(nodes)
G.add_weighted_edges_from(edges)
```

```python
nx.algorithms.shortest_paths.weighted.dijkstra_path_length(G,((0, 0), 't'),(target, 't'),weight='weight')
```
