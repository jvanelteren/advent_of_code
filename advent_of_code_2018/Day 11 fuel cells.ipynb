```python
import numpy as np 
i = 3 
grid_size = (302,302)
#conv_size = (3,3)
grid_serial= 9110

def getpowerlevel(coordinates,grid_serial):
    
    rack_id = coordinates[0]+10
    power_level = rack_id * coordinates[1] 
    result = power_level + grid_serial
    result = result * rack_id
    result = result%(10**i)//(10**(i-1))
    result += -5
    return result



def return_conv_score(cell_grid,coordinates,conv_size):
    return cell_grid[coordinates[0]:coordinates[0]+conv_size[0],coordinates[1]:coordinates[1]+conv_size[1]].sum()

def return_conv(cell_grid,coordinates,conv_size):
    return (cell_grid[coordinates[0]:coordinates[0]+conv_size[0],coordinates[1]:coordinates[1]+conv_size[1]])


def populategridscore(cell_grid,conv_size):#returns grid scores
    grid_scores = np.empty(grid_size)
    grid_scores = np.swapaxes(grid_scores,0,1)
    
    xrange=range(1,int(grid_size[0]-conv_size[0]-1))
    #print ('xrange',xrange)
    yrange=range(1,int(grid_size[0]-conv_size[0]-1))
    #print (yrange)
    for x in xrange :
        for y in yrange:
            grid_scores[x,y]=return_conv_score(cell_grid,(x,y),conv_size)
    #print ('result populategridscores',grid_scores)
    return grid_scores 

def determinemax(grid_scores,conv_size):
    #np.max(grid_scores))
    location = np.where(grid_scores == grid_scores.max())
 
    result =  []
    result.append(grid_scores.max())
    result.append(len(location))
    result.append(location)
    result.append(conv_size)
    #print ('result determinmax (grid_score max, times max, set coordinates, conv_size)',result)

    return result
    #print (cell_grid[21,13])

cell_grid = np.empty(grid_size)
cell_grid = np.swapaxes(cell_grid,0,1)
#populate the grid
for x in range(grid_size[0]):
    for y in range(grid_size[0]):
        cell_grid[x,y]=getpowerlevel((x,y),grid_serial)

        
finalresult = []
```

```python
finalresult = []

def endresult(a,b):
    for x in range(a,b):
        print (x)
        finalresult.append(determinemax(populategridscore(cell_grid,(x,x)),(x,x)))
        
%load_ext line_profiler
%prun endresult(1,301)
                                  
print (finalresult)
    
#print (getpowerlevel((33,45),grid_serial))

#TODO nieuwe numpy array met grid_scores met varierende conv size per fuel_cell
#
```
