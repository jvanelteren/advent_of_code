```python
def getduration(letter):
    return 60+int(str(ord(letter) - 64))

#puzzle2

G = nx.DiGraph()

G.add_edges_from(edges)

result = ''
seconds = 1500
idleworkers = 5
workerready = {}
letterready = {}

for x in range(seconds):
    workerready[x]=0
    letterready[x]=[]

for second in range(seconds):
    print ('start of',second)
    #remove letters that are finished
    if len(letterready[second])>0:
        for letter in letterready[second]:
            G.remove_node(letter)
            print (letter,'removed')
        
        
    if (len(G.nodes())==1):
        print ('ready')
        print (G.nodes()[0],second+getduration(G.nodes()[0]))

    
    idleworkers += workerready[second]
    candidates = G.in_degree(nodes)
    
    #
    potentials = [c for c in candidates.keys() if candidates[c]==0]
    print ('p',potentials)
    truepotential =  [p for p in potentials if p not in result]
    print ('t',truepotential)
    
    
    #
    
    
    
    
    
    while (idleworkers>0) and (len(truepotential)>0):
        print (idleworkers)
        letter = min(truepotential)
        result += letter
        potentials = [c for c in candidates.keys() if candidates[c]==0]
        print ('p',potentials)
        truepotential =  [p for p in potentials if p not in result]
        print ('t',truepotential)      
        candidates = G.in_degree(nodes)          
        idleworkers+=-1

        workerready[second+getduration(letter)]+=1
        letterready[second+getduration(letter)].append(letter)
        print ('letterready',second+getduration(letter),letterready[second+getduration(letter)])

    

```
