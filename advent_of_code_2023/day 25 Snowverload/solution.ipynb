```python
from collections import defaultdict
from aocutils.visuals import visualize_graph
from aocutils.special import UnionFind
import networkx
from math import prod
```

```python
edges = []
f = open("in.txt").read().split("\n")
lines = [line for line in f]
conn = defaultdict(list)
for line in lines:
    l,*r = line.split()
    for n in r:
        edge = {n.strip(), l[:3]}
        if edge not in remove:
            edges.append((l[:3], n.strip()))
```

```python
G = networkx.Graph()
G.add_edges_from(edges)
G.remove_edges_from(networkx.minimum_edge_cut(G))
groups = list(networkx.connected_components(G))
prod(len(g) for g in groups)
```
