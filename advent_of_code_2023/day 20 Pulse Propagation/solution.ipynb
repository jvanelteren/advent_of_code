```python
from collections import defaultdict,deque
from aocutils.math import lcm
```

```python
f = open("in.txt").read().split("\n")
lines = [(line) for line in f]
mod = defaultdict(dict)
end = []
for line in lines:
    left, dests = line.replace(' ', '').split('->')
    name = left[1:]
    mod[name]['dest'] = dests.split(',')
    mod[name]['state'] = 0
    if left.startswith('%'):
        mod[name]['type'] = 'ff'
    elif left.startswith('&'):
        mod[name]['type'] = 'con'
    else:
        mod[name]['type'] = 'broadcaster'
    
for m in list(mod):
    moduletype = mod[m]['type']
    for n in mod[m]['dest']:
        if n in mod and mod[n]['type'] == 'con':
            mod[n][m] = 0
            
cycles = {}
lows, high = 0,0
for step in range(1,11112):
    tocheck = deque([('roadcaster', 0, 'but')])
    while tocheck:
        des, pulse, orig = tocheck.popleft()
        if pulse == 0:
            lows +=1
        if pulse == 1:
            high += 1
        if des not in mod:
            continue
        d, t = mod[des]['dest'], mod[des]['type'] 
        if t == 'ff':
            if pulse == 0:
                mod[des]['state'] ^= 1
                pulse = mod[des]['state']
            else:
                continue
        elif t == 'con':
            mod[des][orig] = pulse
            pulse = 0
            for key in mod[des]:
                if key in mod:
                    if mod[des][key] == 0:
                        pulse = 1
                        break
            if pulse == 0 and des in ['gf', 'vc', 'db', 'qx']:
                if des not in cycles:
                    cycles[des] = step
        for n in d:
            tocheck.append((n, pulse, des))
    if step == 1000: print(lows * high)
lcm(list(cycles.values()))
```

```python
# original
f = open("in.txt").read().split("\n")
lines = [(line) for line in f]
ff = defaultdict(dict)
con = defaultdict(dict)
modules = defaultdict(str)
broad = []
end = []
for line in lines:
    left, right = line.split(' -> ')
    left = left[:4].strip()
    if left.startswith('%'):
        dests = right.replace(' ','').split(',')
        ff[left[1:]]['dest'] = dests
        ff[left[1:]]['state'] = 0
        modules[left[1:]] = 'ff'
    if left.startswith('&'):
        dests = right.replace(' ','').split(',')
        con[left[1:]]['dest'] = dests
        # con[left[1:]]['memo'] = 0
        modules[left[1:]] = 'con'
    if left.startswith('bro'):
        broad = right.replace(' ','').split(',')

    
for m in list(modules):
    moduletype = modules[m]
    if moduletype == 'ff':
        for n in ff[m]['dest']:
            if modules[n] == 'con':
                con[n][m] = 0
            if modules[n] == 'ff':
                con[n][m] = 0
    else:
        assert moduletype == 'con'
        for n in con[m]['dest']:
            if modules[n] == 'con':
                con[n][m] = 0
            if modules[n] == 'ff':
                con[n][m] = 0
    
lows, high = 0,0

for step in range(1,100000):
    tocheck = deque([('bro', 0, 'but')])
    while tocheck:
        des, pulse, orig = tocheck.popleft()
        # print('pulse',des, pulse, orig)
        if pulse == 0:
            lows +=1
        if pulse == 1:
            high += 1
            
        if des == 'bro':
            for n in broad:
                # print('..send broad', n, pulse)
                tocheck.append((n,pulse, des))
        elif modules[des] == 'ff':
            if pulse == 1:
                continue
            else:
                ff[des]['state'] ^= 1
                for n in ff[des]['dest']:
                    # print('.. send ',n,ff[des]['state'])
                    tocheck.append((n,ff[des]['state'], des))
        elif modules[des] == 'con':
            
            con[des][orig] = pulse
            
            allhighs = 1

            for key in con[des]:
                if key != 'dest':
                    if con[des][key] == 0:
                        allhighs = 0
                        break
            if allhighs == 1 and des in ['gf', 'vc', 'db', 'qx']:
                print(des, step)
            
            for n in con[des]['dest']:
                # print('.. send ',n,allhighs^1)
                tocheck.append((n,allhighs^1, des))
        else:
            pass
    # print('cycle over', ff, con)
    for node in ['gf', 'vc', 'db', 'qx']:
        allhighs = 1
        for key in con[node]:
            if key != 'dest':
                if con[node][key] == 0:
                    allhighs = 0
                    break
        if allhighs:
            
            
            print(node, step)
            
    
        
lows *high
```
