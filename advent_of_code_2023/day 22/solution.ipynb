```python
from aocutils.common import ints
f = open("in.txt").read().split("\n")
lines = [line.split('~') for line in f]
alltaken = set()
bricks = []
for l,r in lines:
    a,b,c = ints(l)
    a2,b2,c2 = ints(r)
    if a > a2: a, a2 = a2, a
    if b > b2: b, b2 = b2, b
    if c > c2: c, c2 = c2, c
    taken = set()
    for x in range(a,a2+1):
        for y in range(b,b2+1):
            for z in range(c, c2+1):
                taken.add((x,y,z))
    alltaken |= taken
    bricks.append(taken)
                
def canfall(tile):
    x,y,z = tile
    if z == 1:
        return False
    return (x, y, z-1) not in alltaken

def fall(tile):
    x,y,z = tile
    return (x, y, z-1)

steps = 0
somefalling = True
while somefalling:
    somefalling = False
    steps += 1
    newbricks = []
    for b in bricks:    
        for x,y,z in b:
            if z == 1:
                newbricks.append(b)
                break
        else:
            # not at ground level
            alltaken -= b
            falling  = True
            while falling:
                falling = False
                if all(canfall(tile) for tile in b):
                    falling = True
                    somefalling = True
                    newbrick = set()
                    for tile in b:
                        newbrick.add(fall(tile))
                    b = newbrick
            # print(len(newbrick)                )
            alltaken |= b
            newbricks.append(b)
    bricks = newbricks

```

```python
safe = 0
maxx = 0
total = 0
from copy import deepcopy
savealltaken = deepcopy(alltaken)
savebricks = deepcopy(bricks)
for b in bricks:
    alltaken = deepcopy(savealltaken)
    bricks = deepcopy(savebricks)
    alltaken -= b
    bricks.remove(b)

    somefalling = True
    fallen = set()
    while somefalling:
        somefalling = False
        newbricks = []
        for idx, b2 in enumerate(bricks):    
            for x,y,z in b2:
                if z == 1:
                    newbricks.append(b2)
                    break
            else:
                # not at ground level
                alltaken -= b2
                falling  = True
                while falling:
                    falling = False
                    if all(canfall(tile) for tile in b2):
                        falling = True
                        somefalling = True
                        fallen.add(idx)
                        newbrick = set()
                        for tile in b2:
                            newbrick.add(fall(tile))
                        b2 = newbrick
                # print(len(newbrick)                )
                alltaken |= b2
                newbricks.append(b2)
        bricks = newbricks
    maxx = max(maxx, len(fallen))
    if len(fallen) == 0:
        safe += 1
    total += len(fallen)
alltaken = deepcopy(savealltaken)
bricks = deepcopy(savebricks)

safe, maxx + 1, total
```
