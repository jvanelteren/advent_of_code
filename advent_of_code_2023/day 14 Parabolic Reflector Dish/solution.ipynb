```python
from aocutils.special import find_cycle
from aocutils.grid import iterate
import numpy as np

f = open("in.txt").read().split("\n")
arr = [list(line) for line in f]
rows, cols = len(arr), len(arr[0])


def fall(row, rev=False):
    for i in range(len(row)):
        if row[i] == '.':
            cur = i + 1
            while 0 <= cur < len(row):
                if row[cur] == '#':
                    break
                elif row[cur] == 'O':
                    row[i] = 'O'
                    row[cur] = '.'
                    break
                cur += 1
    return row


def north(arr):
    for r in range(rows):
        for c in range(cols):
            val = arr[r][c]
            if val == '.':
                cur = r
                while cur < len(arr):
                    if arr[cur][c] == 'O':
                        arr[r][c] = 'O'
                        arr[cur][c] = '.'
                        break
                    elif arr[cur][c] == '#':
                        break
                    else:
                        cur += 1
    return arr





def south(arr):
    for r in range(rows-1,-1,-1):
        for c in range(cols):
            val = arr[r][c]
            if val == '.':
                cur = r
                while cur > -1:
                    if arr[cur][c] == 'O':
                        arr[r][c] = 'O'
                        arr[cur][c] = '.'
                        break
                    elif arr[cur][c] == '#':
                        break
                    else:
                        cur -= 1
    return arr
def east(arr):
    for c in range(cols-1,-1,-1):
        for r in range(rows):
            val = arr[r][c]
            if val == '.':
                cur = c
                while cur > -1:
                    if arr[r][cur] == 'O':
                        arr[r][c] = 'O'
                        arr[r][cur] = '.'
                        break
                    elif arr[r][cur] == '#':
                        break
                    else:
                        cur -= 1
    return arr

def west(arr):
    for c in range(cols):
        for r in range(rows):
            val = arr[r][c]
            if val == '.':
                cur = c
                while cur < cols:
                    if arr[r][cur] == 'O':
                        arr[r][c] = 'O'
                        arr[r][cur] = '.'
                        break
                    elif arr[r][cur] == '#':
                        break
                    else:
                        cur += 1
    return arr
tup = lambda arr: tuple(tuple(r) for r in arr)
lis = lambda arr: list(list(r) for r in arr)
def cycle(arr):
    arr = lis(arr)
    arr = north(arr)
    arr = west(arr)
    arr = south(arr)
    arr = east(arr)
    return tup(arr)
    

def count(arr):
    ans = 0
    for r,c, val in iterate(arr):
        if val == 'O':
            ans += rows - r
    return ans

for i in range(20+140):
    arr = cycle(arr)
count(arr)
```
