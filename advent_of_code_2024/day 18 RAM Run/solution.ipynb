```python
from collections import deque
from aocutils.common import ints
from aocutils.grid import move

E = (0, 1)
W = (0, -1)
N = (-1, 0)
S = (1, 0)
winds = [N, E, S, W]
```

```python
# did manual binary search, although bruteforce would also work
f = open("in.txt").read().split("\n")
lines = [ints(line) for line in f]
maxr = 71
maxc = maxr

def neigh(loc):
    options = []
    for wind in winds:
        newloc = move(loc, wind)
        if 0 <= newloc[0] < maxr and 0 <= newloc[1] < maxc and newloc not in forbid:
            options.append(newloc)
    return options

forbid = set()

for idx in range(1024):
    forbid.add(lines[idx])

start = (0, 0)
goal = (maxr - 1, maxc - 1)

seen = set()
dis = 0
tocheck = deque([start])
while tocheck:
    for _ in range(len(tocheck)):
        cur = tocheck.popleft()
        for n in neigh(cur):
            if n not in seen:
                seen.add(n)
                tocheck.append(n)
                if n == goal:
                    print(lines[idx], dis + 1)
                    break
    dis += 1
```

```python
# version with libraries
from aocutils.special import binarysearch
from aocutils.maze import bfs


def runbfs(i):

    forbid = set()
    for idx in range(i):
        forbid.add(lines[idx])

    def neigh(loc):
        options = []
        for wind in winds:
            newloc = move(loc, wind)
            if 0 <= newloc[0] < maxr and 0 <= newloc[1] < maxc and newloc not in forbid:
                options.append(newloc)
        return options

    return bfs(neigh, start, goal)


binarysearch(0, len(lines), runbfs, flips_to_true=False)
```
