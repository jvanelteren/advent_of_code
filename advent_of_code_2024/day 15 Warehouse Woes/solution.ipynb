```python
f = open("in.txt").read().split("\n\n")
lines = [line for line in f]
grid, ins = lines
ins = ins.replace('\n','')
newgrid = ''
for ch in grid:
    newgrid += {'#':'##', '.':'..', '@':'@.', 'O':'[]', '\n':'\n'}[ch]
grid = newgrid
grid = arr_to_dict([list(line) for line in grid.split('\n')])
neigh = arr_neighbors([list(line) for line in grid])
start = [k for k,v in grid.items() if v == '@'].pop()
changedir = {E:W, W:E, N:S, S:N}

def switch(a,b):
    grid[a], grid[b] = grid[b], grid[a]

for i in ins:
    direc = {'v':S, '<':W, '>':E, '^':N}[i]
    cur = move(start, direc)
    if grid[cur] == '.': # just move the robot
        switch(start, cur)
        start = cur
        continue
    elif grid == '#': # we cant move
        continue
    elif direc in [E,W]: # this is easier than up or down
        while grid[cur] in '[]O':
            cur = move(cur, direc)
        if grid[cur] == '.':
            newcur = move(cur, changedir[direc])
            while grid[newcur] != '@': # go back 
                switch(cur, newcur)
                cur = newcur
                newcur = move(cur, changedir[direc])
            switch(cur, newcur) # one more time to 
            start = cur
    elif direc in [N,S]:
        moving = set()
        tocheck = {start}
        wall = False
        while tocheck:
            box = tocheck.pop()
            newloc = move(box, direc)
            if grid[newloc] == '#':
                wall = True
                break
            elif grid[newloc] in '[]':
                tocheck.add(newloc)
                moving.add(newloc)
                friendbox = newloc[0], newloc[1]+ {'[':+1, ']':-1}[grid[newloc]]
                tocheck.add(friendbox)
                moving.add(friendbox)
        if wall: continue
        moving = sorted(list(moving), reverse = {N:False, S:True}[direc])        
        for box in moving:
            newcur = move(box, direc)
            switch(box, newcur)
        switch(start, move(start, direc))
        start = move(start, direc)
sum([100*k[0] + k[1] for k,v in grid.items() if v == '['])
```
